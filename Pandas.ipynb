```python
dts=pd.read_csv("http://bit.ly/imdbratings")
```

```python
#All the movies with duration greater than 200 minutes
x=dts['duration']>200
dts[x]
```

```python
#All the movies with duration greater than 200 minutes AND belong to the genre Drama
dts[(dts['duration']>200) & (dts['genre']=='Drama')]
```

```python
#All the movies with duration greater than 200 minutes AND belong to the genre Drama OR Crime
dts[(dts['duration']>200) & ((dts['genre']=='Drama')|(dts['genre']=='Crime'))]
```

```python
#All the movies with duration greater than 200 minutes AND belong to the genre Drama or Crime or Action
dts[(dts['duration']>200) & (dts['genre'].isin(['Drama','Crime','Action']))]
```

```python
#All the movies with actors that have Aamir or Shah in their names
dts[(dts.actors_list.str.contains('Aamir'))|(dts.actors_list.str.contains('Shah'))]
```

```python
#All the movies with actors that have Khan in their names
dts[(dts.actors_list.str.contains('Khan'))]
```
