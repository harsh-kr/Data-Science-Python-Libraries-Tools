```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use('ggplot')
```

```python
x=np.linspace(1,10)
y=np.sin(x)
plt.plot(x,y,':',label='sin')
plt.plot(x,np.cos(x),'--',label='cos')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("plots")
plt.legend()
#plt.savefig('g:/my.png')
plt.show()
```

```python
x=[1,3,2,4,6,5]
y=[7,3,4,7,9,8]
plt.scatter(x,y,marker='X',s=150,alpha=1,label='sactter',color='blue')
plt.legend(loc='upper left')
plt.show()
```

```python
x=[1,3,2,12,23,12,24,14,15,16,17,14,18,21,12,12,23,24,25,26,4,3,2,1,1]
r=[1,10,15,30]
plt.hist(x,r,histtype='bar',rwidth=0.1)
plt.show()
```

```python
x=['a','b','c','d','e']
x1=list(range(len(x)))
y=[9,2,5,4,7]
plt.bar(x1,y)#tick_label=x)
#plt.xticks(x1,x)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,6,2,8,7,4]
y1=[5,8,9,7,2,4]
y2=[1,2,5,7,9,3]
d=dict(colors=['blue','green','yellow'],labels=['data1','data2','data3'])

plt.stackplot(x,y,y1,y2,**d)

plt.legend()
plt.show()
```

```python
x=['a','b','c','d','e']
y=[9,2,5,4,7]
plt.pie(y,autopct='%0.2f',labels=x,explode=(0.1,0,0,0.2,0),shadow=True)
plt.show()
```
